```python




import pandas as pd
loan=pd.read_csv("C:\\Users\\spoor\\Downloads\\loan_approved.csv")
df=pd.DataFrame(loan)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
loan_enc=le.fit_transform(df['Loan_ID'])
print(loan_enc)
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
gender_enc=le.fit_transform(df['Gender'])
print(gender_enc)
gender=pd.get_dummies(df['Gender'])
print(gender)
```

```python
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
married_enc=le.fit_transform(df['Married'])
print(married_enc)
```

```python
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
enc1=le.fit_transform(df['Dependents'])
print(enc1)
enc11=pd.get_dummies(df['Dependents'],dtype=int)
print(enc11)
```

```python
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
self=le.fit_transform(df['Self_Employed'])
print(self)
```

```python
appinc=pd.get_dummies(df['ApplicantIncome'])
print(appinc)
```

```python
coappl=le.fit_transform(df['CoapplicantIncome'])

coapp=pd.get_dummies(df['CoapplicantIncome'],dtype=int)
print(coapp)
```

```python
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].mean())
```

```python
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0])
loanterm=pd.get_dummies(df['Loan_Amount_Term'],dtype=int)
print(loanterm)
```

```python
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mode()[0])
cred=pd.get_dummies(df['Credit_History'],dtype=int)
print(cred)
```

```python
area=pd.get_dummies(df['Property_Area'],dtype=int)
print(area)
area1=le.fit_transform(df['Property_Area'])
print(area1)
```

```python
df.head()
```

```python
categorical_cols = ['Loan_ID','Gender','Married','Dependents','Education',
                    'Self_Employed','Property_Area','Loan_Status (Approved)']

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

print(df.head())
```
